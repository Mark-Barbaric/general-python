# @property Decorators Workbook

```python
class PropertyClass(object):
    def __init__(self):
        self._count = None
```

```python
property_class = PropertyClass()
property_class._count
```

```python
property_class._count = 22
property_class._count
```

This could cause some problems. Now we are going to use the @property decorate to restrict how this class attribute can be edited. Now we will redefine the class with a private parameter called count, and use the @property decorator to set it as the getter.

```python
class PropertyClassWithoutSetter(object):
    def __init__(self):
        self._count = None

    @property
    def count(self):
        return self._count
```

```python
property_class_without_setter = PropertyClassWithoutSetter()
property_class_without_setter.count
```

Trying to change the value of the count directly and using the setter which hasn't been defined yet.

```python
try:
    property_class_without_setter.count = 11
except AttributeError as error:
    print(f"AttributeError: {error}")
except NameError as error:
    print(f"NameError: {error}")
```

```python
try:
    property_class_without_setter._count = 11
except NameError as error:
    print(f"NameError: {error}")
finally:
    print(property_class_without_setter.count)
```

```python
class PropertyClassWithSetter(object):
    def __init__(self):
        self._count = None

    @property
    def count(self):
        return self._count

    @count.setter
    def count(self, new_count: int):
        if new_count < 0 or new_count > 22:
            raise ValueError
        self._count = new_count
```

```python
property_class_with_setter = PropertyClassWithSetter()
property_class_with_setter.count
```

```python
property_class_with_setter.count = 22
property_class_with_setter.count
```

```python
property_class_with_setter._count = -22
```
